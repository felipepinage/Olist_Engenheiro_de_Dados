```
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
path1 = "s3a://bucket-olist-pinage/olist_sellers_dataset.csv"

sellers_df = (spark
  .read
  .option("header", True)
  .option('delimiter', ',')
  .csv(path1)
)

#display(sellers_df)
```

```
path2 = "s3a://bucket-olist-pinage/olist_customers_dataset.csv"

cust_df = (spark
  .read
  .option("header", True)
  .option('delimiter', ',')
  .csv(path2)
)

#display(cust_df)
```

```
path3 = "s3a://bucket-olist-pinage/olist_orders_dataset.csv"

orders_df = (spark
  .read
  .option("header", True)
  .option('delimiter', ',')
  .csv(path3)
)

#display(order_df)
```

```
path4 = "s3a://bucket-olist-pinage/olist_order_items_dataset.csv"

items_df = (spark
  .read
  .option("header", True)
  .option('delimiter', ',')
  .csv(path4)
)

#display(items_df)
```

```
dim_customers = cust_df.select(col('customer_id'), col('customer_unique_id'))
#display(dim_customers)
```

```
ship_date_df = items_df.select(col('order_id'), col('shipping_limit_date'))

dim_order_date = orders_df.select(col('order_id'),\
                                  col('order_purchase_timestamp'),\
                                  col('order_approved_at'),\
                                  col('order_delivered_carrier_date'),\
                                  col('order_delivered_customer_date'),\
                                  col('order_estimated_delivery_date')).\
                                  join(ship_date_df, 'order_id')
#display(dim_order_date)
```

```
dim_order_date_dropdup = dim_order_date.drop_duplicates(subset=['order_id'])
```

```
#display(items_df)
```

```
# Para incluir na tabela fact_items_sales
order_cust_df = orders_df.select(col('order_id'), col('order_status'), col('customer_id'))
cust_zip_status_df = cust_df.select(col('customer_id'), col('customer_zip_code_prefix'))\
                         .join(order_cust_df, 'customer_id')
#display(cust_zip_status_df)
```

```
fact_item_sales_tst = items_df.join(cust_zip_status_df, 'order_id').drop(col('shipping_limit_date'))
fact_item_sales = fact_item_sales_tst.join(sellers_df, 'seller_id').drop(col('seller_city')).drop(col('seller_state'))
#display(fact_item_sales)
```

```
dim_customers.coalesce(1).write \
    .format('csv') \
    .option("header", True) \
    .option("delimiter" , "|") \
    .save('s3a://bucket-olist-pinage/olist_dim_customers_dataset')

dim_sellers.coalesce(1).write \
    .format('csv') \
    .option("header", True) \
    .option("delimiter" , "|") \
    .save('s3a://bucket-olist-pinage/olist_dim_sellers_dataset')

dim_order_date_dropdup.coalesce(1).write \
    .format('csv') \
    .option("header", True) \
    .option("delimiter" , "|") \
    .save('s3a://bucket-olist-pinage/olist_dim_order_date_dataset')

fact_item_sales.coalesce(1).write \
    .format('csv') \
    .option("header", True) \
    .option("delimiter" , "|") \
    .save('s3a://bucket-olist-pinage/olist_fact_item_sales_dataset')
```

```
path5 = "s3a://bucket-olist-pinage/olist_order_payments_dataset.csv"
path6 = "s3a://bucket-olist-pinage/olist_order_reviews_dataset.csv"
path7 = "s3a://bucket-olist-pinage/olist_products_dataset.csv"
path8 = "s3a://bucket-olist-pinage/olist_geolocation_dataset.csv"

pay_df = (spark.read.option("header", True).option('delimiter', ',').csv(path5))
reviews_df = (spark.read.option("header", True).option('delimiter', ',').csv(path6))
prod_df = (spark.read.option("header", True).option('delimiter', ',').csv(path7))
geo_df = (spark.read.option("header", True).option('delimiter', ',').csv(path8))
```

```
fact_order_payments = pay_df.join(order_cust_df, 'order_id').drop(col('order_status'))

fact_order_reviews = reviews_df.join(order_cust_df, 'order_id').drop(col('order_status'))

dim_geolocation_dropdup = geo_df.drop_duplicates(subset=['geolocation_zip_code_prefix'])
```

```
fact_order_payments.coalesce(1).write \
    .format('csv') \
    .option("header", True) \
    .option("delimiter" , "|") \
    .save('s3a://bucket-olist-pinage/olist_fact_order_payments_dataset')

fact_order_reviews.coalesce(1).write \
    .format('csv') \
    .option("header", True) \
    .option("delimiter" , "|") \
    .save('s3a://bucket-olist-pinage/olist_fact_order_reviews_dataset')

dim_products.coalesce(1).write \
    .format('csv') \
    .option("header", True) \
    .option("delimiter" , "|") \
    .save('s3a://bucket-olist-pinage/olist_dim_products_dataset')

dim_geolocation_dropdup.coalesce(1).write \
    .format('csv') \
    .option("header", True) \
    .option("delimiter" , "|") \
    .save('s3a://bucket-olist-pinage/olist_dim_geolocation_dataset')
```
